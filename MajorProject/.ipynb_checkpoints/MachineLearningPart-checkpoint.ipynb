## Creating a logistic regression to predict the Absenteeism

### Importing the Libraries

```python
import pandas as pd
import numpy as np
```

### Load the data

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
data_preprocessed.head()
```

### Logistic Regression - It is a type of classification where we will be classifying the people into the classes

### There are two types of classes present in this scenario

#### 1. Excessively Absent
#### 2. Moderately Absent

#### Note : In order to calculate the actual class we need to calculate the median value of the Absenteeism Time in Hours, this will classify the actual class of the scenrio

### Creating the target

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

### If the person is absent for more than 3 Hours , then it is considered to be as the excessively absent. Similarly less than that are considered as the moderately absent.

#### Anything above the medium is excessive absent and anything below the medium is moderate absent.

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] >
                   data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed
```

### A comment on the targets

```python
targets.sum()/targets.shape[0]
```

#### The above results clearly shows that almost 45% targets are 1s and rest 54% are 0s.

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets
```

```python
# using the reserve word here and since it shows that True - which means that the two variables refer to the same object 
# and since it shows the False which means that the two variables are of the different objects.
data_with_targets is data_preprocessed
```

### Select the input for the regression

```python
data_with_targets.shape
```

```python
unscaled_input = data_with_targets.iloc[:,:-1]
```

```python
unscaled_input
```

## Standardize the data

```python
from sklearn.preprocessing import StandardScaler
abseentism_scaler = StandardScaler()
```

```python
abseentism_scaler.fit(unscaled_input)
```

```python
scaled_input = abseentism_scaler.transform(unscaled_input)
```

```python
scaled_input
```

```python
scaled_input.shape
```

## Split the data into train & test and shuffle

### Import the relevant module

```python
from sklearn.model_selection import train_test_split
```

### Split

```python
train_test_split(scaled_input,targets)
```

```python
x_train , x_test , y_train , y_test = train_test_split(scaled_input,targets,train_size = 0.8,random_state = 20)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

### Here , 75 percent of the observation will help us with training and 25 percent will serve for testing

### Logistic Regression for Sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

### Based on the present data, our model understands the data and learn to classify 80% of the observations correctly.

```python
reg.score(x_train,y_train)
```

### Manually check the accuracy

```python
model_output = reg.predict(x_train)
model_output
```

```python
model_output == y_train
```

```python
print("The number of the correct predictions are : " , np.sum(model_output==y_train))
```

```python
model_output.shape[0]
```

```python
np.sum(model_output==y_train) / model_output.shape[0]
```

### Find the intercept and the coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_input.columns.values
```

```python
feature_name = unscaled_input.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'],data=feature_name)

summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.iloc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

### Interpreing the coefficients

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('Odds_ratio')
```
